### Natural Language Processing
#### KLTK 
KLTK는 알고리즘 기반의 한국어 형태소 분석기인 "Korean Language Toolkit"의 약자입니다.

KLTK는 텍스트 데이터를 형태소 단위로 분석하여 문장의 구조를 이해하고, 각 단어의 품사와 의미를 파악하는 역할을 합니다. 
 
이를 통해 한국어 텍스트의 자동화 처리, 자연어 처리, 정보 검색, 기계 번역 등 다양한 언어 관련 작업에 사용될 수 있습니다.

KLTK는 형태소 분석과 관련된 다양한 기능을 제공하며, 한국어 문장을 입력으로 받아서 형태소로 분리하고 각 형태소의 품사를 태깅해줄 수 있습니다. 

이를 통해 텍스트 데이터를 의미 있는 단위로 분할하여 정보 추출이나 다른 자연어 처리 작업에 활용할 수 있습니다.

KLTK는 한국어 자연어 처리에 유용한 도구 중 하나이며, 학술 연구나 산업적인 응용 분야에서 활용되고 있습니다.

### WordCloud : 워드 클라우드
텍스트 데이터에서 단어들의 빈도를 시각적으로 나타내는 방식입니다. 

주어진 텍스트에서 각 단어의 출현 빈도를 계산하고, 빈도가 높은 단어일수록 시각적으로 크게 표현됩니다. 

워드 클라우드는 텍스트 데이터에서 주요한 키워드를 한 눈에 파악할 수 있는 시각화 도구로 사용됩니다.

### 1. 이상한 나라의 앨리스 

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
import requests
from PIL import Image
from io import BytesIO


test_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/nltk_dataset/06_alice.txt'
image_url = 'https://github.com/PinkWink/ML_tutorial/blob/master/nltk_dataset/06_alice_mask.png?raw=true'

response = requests.get(test_url)
text = response.text

response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
alice_mask = np.array(image)
```

```python
print(text)
```

```python
alice_mask 
```

<b>stopwords</b>

WordCloud 객체에서 stopwords 매개변수는 워드 클라우드를 생성할 때 무시할 단어들의 목록을 지정하는 데 사용됩니다. 

이렇게 지정된 stopwords를 제외하고 나머지 단어들을 워드 클라우드에 포함시킨다.

```python
stopwords = set(STOPWORDS)
stopwords
```

```python
stopwords.add('said')
```

```python
# %load ../set_matplotlib_hangul.py

import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
import platform

# %matplotlib inline
get_ipython().run_line_magic("matplotlib","inline")

path='C:/Windows/Fonts/malgun.ttf'
plt.rcParams['axes.unicode_minus'] = False

if platform.system()=="Darwin": #mac
    rc("font",family='Arial Unicodes MS')
    sns.set(font='Arial Unicodes MS', 
        rc={"axes.unicode_minus":False}),
    print('MAC Hangul OK')
elif platform.system()=="Windows":  #window
    font_name=font_manager.FontProperties(fname=path).get_name()
    rc("font",family=font_name)
    sns.set(font=font_name, 
    rc={"axes.unicode_minus":False})
    print('WIndow Hangul OK')
    
else:
    print('Unknown System')
    
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.grid(False)
```

```python
wc = WordCloud(
    background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords #
)

wc = wc.generate(text)
```

```python
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

### 3. 육아 휴직 관련 법안 

```python
import nltk
from konlpy.corpus import kobill
```

```python
#kobill 말뭉치는 대한민국 국회에서 제공하는 법률 문서들로 구성되어 있습니다. 
#이 말뭉치는 다양한 법률 관련 텍스트 데이터를 제공하여 한국어 자연어 처리 작업에 활용할 수 있습니다. 
doc_ko = kobill.open('1809890.txt').read()
```

```python
print(doc_ko)
```

```python
from konlpy.tag import Okt
```

```python
t = Okt()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='육아휴직')
ko
```

```python
len(ko.tokens) # 명사는 735개 
```

```python
len(set(tokens_ko)) #유일한 단어들 
```

```python
ko.vocab()
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 1809890호')

plt.figure(figsize=(12, 6))
ko.plot(50) #Plot sorted frequency of top 50 tokens
plt.show()
```

```python
stop_words = [
    ".",
    "(",
    ")",
    ",",
    "'",
    "%",
    "-",
    "X",
    ").",
    "x",
    "의",
    "자",
    "에",
    "안",
    "번",
    "호",
    "을",
    "이",
    "다",
    "만",
    "로",
    "가",
    "를"
]
```

```python
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 1809890호')

plt.figure(figsize=(12, 6))
ko.plot(50) #Plot sorted frequency of top 50 tokens
plt.show()
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(12, 6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원']) # 각 단어들이 어디쯤 위치했는지
```

**영어에서 연어(collocation)란 무엇인가?**

연어란 "함께 위치하는 단어들"(co+location)이란 뜻으로, 어휘의 조합 또는 짝을 이루는 말들을 일컫습니다. 

ex - 'friend' 와 주로 같이 사용되는 단어들 : 'best', 'good'

```python
ko.collocations()
```

```python
data = ko.vocab().most_common(200)

path='C:/Windows/Fonts/malgun.ttf'

wordcloud = WordCloud(
    font_path=path,
    relative_scaling=0.2, #간격
    background_color='white'
).generate_from_frequencies(dict(data))

plt.figure(figsize=(10,8), dpi=100)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

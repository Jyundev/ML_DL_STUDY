```python
def log(x, base):
    return np.log(x)/np.log(base)


x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.01, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)
y21, y22 = log(x1, 1/10), log(x2, 1/np.e)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(x1, y11, label='$\log_{10} x$', color='k')
ax[0].plot(x2, y12, '--', label='$\log_{e} x$', color='k')

ax[0].set_xlabel('$x$', fontsize=20)
ax[0].set_ylabel('$y$', fontsize=20)
ax[0].legend(fontsize=20, loc='lower right')

ax[1].plot(x1, y21, label='$\log_{1/10} x$', color='k')
ax[1].plot(x2, y22, '--', label='$\log_{1/e} x$', color='k')

ax[1].set_xlabel('$x$', fontsize=20)
ax[1].set_ylabel('$y$', fontsize=20)
ax[1].legend(fontsize=20, loc='upper right')
```

#### 시그모이드
시그모이드 함수는 입력값을 0과 1 사이의 값으로 압축하는 비선형 함수입니다.<p>
주로 로지스틱 회귀와 같은 이진 분류 모델에서 활성화 함수로 사용됩니다.<p>
시그모이드 함수는 x가 음수로 갈수록 0에 가까워지고, x가 양수로 갈수록 1에 가까워집니다. 중간값인 x=0일 때는 f(x) = 0.5가 됩니다

```python
z = np.linspace(-10,10,100)
sigma = 1/(1+np.exp(-z))

plt.figure(figsize=(6,6))
plt.plot(z, sigma, '--', color = 'k', label = '$x^2$')
plt.xlabel('$z$', fontsize=20)
plt.ylabel('$\sigma(z)$',fontsize=20)

```

#### meshgrid
meshgrid 함수는 주어진 벡터를 기반으로 모든 가능한 좌표 쌍을 생성합니다. <p>
생성된 좌표 쌍은 각각 2개의 2차원 배열을 반환합니다. <p>
첫 번째 배열은 X 좌표를 포함하고 두 번째 배열은 Y 좌표를 포함합니다.<p>

```python
U, V = np.meshgrid(u, v)
print(U[0]) # 행방향
print(V[:,0]) # 열방향
```

```python
X = U
Y = V
Z = (1+U**2)+V/(1+V**2)
Z
```

```python
fig = plt.figure(figsize=(10, 7))
#ax = plt.axes(projection='3d')
ax = fig.add_subplot(111, projection='3d')


ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

ax.scatter3D(X, Y, Z, marker='.', c='gray')
plt.show()
```

### 함수의 합성
<img src = 'https://cdn1.byjus.com/wp-content/uploads/2021/05/composite-functions.png' alt = 'https://byjus.com/maths/composition-of-functions/'>

```python
x = np.linspace(-4, 4, 100)
y = x**3 - 15*x+30  # f(x)
z = np.log(y)  # g(f(x))
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].plot(x,y,label='$x^3 -15x +30$',c='k')
ax[0].legend(fontsize=20)
ax[1].plot(y,z,label='$\log(y)$',c='k')
ax[1].legend(fontsize=20)

```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].plot(x,z,label='$\log(f(x))$',ls='--',c='k')
ax[0].legend(fontsize=18)

ax[1].plot(x,y,label='$x^3 -15x +30$',c='k')
ax[1].legend(fontsize=18)

ax_tmp=ax[1].twinx()
ax_tmp.plot(x,z,label='$\log(f(x))$',ls='--',c='k')
```

### Box Plot

```python
samples = [1, 7, 9, 16, 36, 39, 45, 45, 46, 48, 51, 100, 101]
```

```python
tmp_y = [1] * len(samples)
tmp_y
```

```python
plt.figure(figsize=(12, 4))
plt.scatter(x=samples, y=tmp_y)
plt.show()
```

#### 지표 찾기

```python
np.median(samples) #중간값
```

```python
np.percentile(samples, 75) - np.percentile(samples, 25) #IQR
```

```python
q1 = np.percentile(samples, 25)
q2 = np.median(samples)
q3 = np.percentile(samples, 75)

iqr = q3-q1

upper_fence = q3+iqr*1.5
lower_fence = q1-iqr*1.5
```

```python
plt.figure(figsize=(12, 4))
plt.scatter(x=samples, y=tmp_y)
plt.axvline(x=q1, c='b')
plt.axvline(x=q2, c='r')
plt.axvline(x=q3, c='b')
plt.axvline(x=upper_fence, ls='--', c='k')
plt.axvline(x=lower_fence, ls='--', c='k')

plt.show()
```

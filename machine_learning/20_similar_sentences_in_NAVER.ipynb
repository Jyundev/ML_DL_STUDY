```python
def getSearchUrl(apiNode, searchText, startNum, dispNum):
    base_url = "https://openapi.naver.com/v1/search"
    node="/"+apiNode+".json"
    param_query="?query="+urllib.parse.quote(searchText)
    param_start="&start="+str(startNum)
    param_disp="&display="+str(dispNum)
    
    return base_url+node+param_query+param_start+param_disp
```

```python
import json
import datetime

def getResultPage(url, client_id, client_secret):
    request=urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    print("[%s] Url Request Success"%datetime.datetime.now())
    rescode = response.getcode()
    
    if(rescode==200):
        return json.loads(response.read().decode('utf-8'))
    else:
        return print("Error Code:" + rescode)
```

```python
def deleteTag(input_str):
    input_str=input_str.replace("<b>","")
    input_str=input_str.replace("</b>","")    
    return input_str
```

```python
def get_description(pages):
    contents = []

    for sentences in pages['items']:
        contents.append(deleteTag(sentences['description']))

    return contents
```

```python
contents = get_description(one_result)
contents
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import Okt
```

```python
t = Okt()
vectorizer = CountVectorizer(min_df=1)
```

```python
contents_tokens = [t.morphs(row) for row in contents]
contents_tokens
```

```python
len(contents_tokens)
```

```python
contents_for_vectorize = []

for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word

    contents_for_vectorize.append(sentence)

contents_for_vectorize
```

```python
X = vectorizer.fit_transform(contents_for_vectorize)
```

```python
num_samples, num_features = X.shape
num_samples, num_features
```

```python
new_post = ['파이썬을 기본서 추천']
new_post_tokens = [t.morphs(row) for row in new_post]

new_post_for_vectorize = []

for content in new_post_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word

    new_post_for_vectorize.append(sentence)

new_post_for_vectorize    
```

```python
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
import scipy as sp

def dist_raw(v1, v2):
    delta = v1-v2
    return sp.linalg.norm(delta.toarray())
```

```python
dist = [dist_raw(each, new_post_vec) for each in X]
dist
```

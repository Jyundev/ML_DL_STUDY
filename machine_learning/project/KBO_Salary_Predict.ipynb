#### 2. 1983년 - 1988년 투수 데이터 가져오기

```python
import pandas as pd
import numpy as np
```

```python
# 페이지별 데이터 가져오기 
def get_current_page_data(driver, year):
    
    req=driver.page_source
    soup=BeautifulSoup(req,'html.parser')        
    kbo_info=soup.select('#resultListDiv > table > tbody > tr')

    #열 데이터
    columns = [col.text for col in kbo_info[0].select('th') ]
    columns.append('YEAR')


    # 행 데이터
    data_rows = kbo_info[1:]
    data = []

    for row in tqdm(data_rows):
        row_data = [data.text for data in row if data.text != '\n']
        row_data.append(year)
        data.append(row_data)
    



    return columns, data
    
```

```python
def get_KBO_data(start_year, end_year, data_type):
    
    # data_type : main - 1, basic - 2, detail - 3  

    type_btn = driver.find_element(By.XPATH, f'//*[@id="container"]/div[9]/div[2]/div/div[4]/a[{data_type}]/div')
    driver.execute_script("arguments[0].click();", type_btn)

    time.sleep(2)

    
    # 연도별 데이터 첫 번째 페이지부터 마지막 페이지 데이터 획득 

    year = start_year
    values=[]

    while tqdm(year <=end_year):

        # season select
        season_select = driver.find_element(By.CSS_SELECTOR,'#searchFrm > div.ltb-select-left > div.ltb-sl-1 > select.ltb-season-select')
        season_select.send_keys(year)

        # 첫번째 페이지로 이동 
        first_page = driver.find_element(By.CSS_SELECTOR, '#paging > a.paging-num.first_page')
        driver.execute_script("arguments[0].click();", first_page)
        
        time.sleep(2)
        
        #결과 버튼 클릭 
        result_btn = driver.find_element(By.CSS_SELECTOR,'#searchFrm > div.ltb-select-right > button')
        result_btn.click()

        time.sleep(2)

        # 아래로 스크롤 이동
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        print(year)
        time.sleep(3)

        # 페이지에서 데이터 수집 수행 - 첫번째 페이지
        col, data = get_current_page_data(driver, year)
        values.append(data)

        while ('#none' in driver.find_element(By.CSS_SELECTOR,'#paging > a.paging-num.last_page').get_attribute("href")) == False :

            # 아래로 스크롤 이동
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # 다음 페이지로 이동
            next_button = driver.find_element(By.CSS_SELECTOR, "#paging > a.paging-num.next_page > i")
            driver.execute_script("arguments[0].click();", next_button)
            
            time.sleep(5)

            # 페이지에서 데이터 수집 수행
            col, data = get_current_page_data(driver, year)

            values.append(data)

        year+=1

    return col, values

        
```

```python
# 메인 기록
print('------------------main------------------')
columns, values = get_KBO_data(1983, 1988, 1)
item = [item for row in values for item in row]
df_main = pd.DataFrame( item , columns=columns)

# 기본 기록
print('------------------basic------------------')
columns, values = get_KBO_data(1983, 1988, 2)
item = [item for row in values for item in row]
df_basic = pd.DataFrame( item , columns=columns)

# 세부 기록
print('------------------detail------------------')
columns, values = get_KBO_data(1983, 1988, 3)
item = [item for row in values for item in row]
df_detail = pd.DataFrame( item , columns=columns)
```

```python
#1. STATIZ 웹사이트 불러오기
from selenium.common.exceptions import NoSuchElementException # 광고배너 에러 방지 
from selenium.webdriver.common.keys import Keys

def loadStatizWeb(driver, url):
    
    # 화면 최대 크기 설정 
    driver.maximize_window()
    driver.get(url)

    # 광고 배너 제거 
    try:
        ad_close = driver.find_element(By.CSS_SELECTOR,'#foin_close_02eu')
        driver.execute_script("arguments[0].click();", ad_close)
        ad_close.click()
        
    except NoSuchElementException:
        print("광고가 존재하지 않습니다.")

    # 스크롤 내리기 
    actions = ActionChains(driver)
    actions.send_keys(Keys.PAGE_DOWN).perform()

```

```python
url="http://www.statiz.co.kr/salary.php"
driver=webdriver.Chrome('chromedriver.exe')
loadStatizWeb(driver, url)
```

```python
def get_salary():
    req=driver.page_source
    soup=BeautifulSoup(req,'html.parser')        
    salary_info=soup.select('table.table.table-striped tr')
    salary_info[0].select('th')

    # 열 데이터 
    columns = [col.text for col in salary_info[0].select('th') ]

    # 행 데이터
    data_rows = salary_info[1:]
    data = []

    for row in tqdm(data_rows):
        row_data = [data.text for data in row]
        data.append(row_data)

    return columns, data

    
```

```python
def get_kbp_salary_data(start_year, end_year):

    year = start_year
    salary_data = []
    
    # 2015 - 35번째 , 2020 - 40번째
    cnt = 35 

    while year <=end_year:

        # 연도 설정 
        date_btn = driver.find_element(By.CSS_SELECTOR, 'body > div.wrapper > div.content-wrapper > div > section.content > div > div:nth-child(2) > div:nth-child(1) > div > div:nth-child(1) > div > button')
        date_btn.click()

        date_select_btn = driver.find_element(By.CSS_SELECTOR, f'body > div.wrapper > div.content-wrapper > div > section.content > div > div:nth-child(2) > div:nth-child(1) > div > div:nth-child(1) > div > ul > div > button:nth-child({cnt})')
        date_select_btn.click()

        time.sleep(2)

        # 연도 선택 시 광고가 다시 뜸 
        try:
            ad_close = driver.find_element(By.CSS_SELECTOR,'#foin_close_02eu')
            driver.execute_script("arguments[0].click();", ad_close)
            ad_close.click()
                
        except NoSuchElementException:
            print("광고가 존재하지 않습니다.")


        # 스크롤 내리기 
        actions = ActionChains(driver)
        actions.send_keys(Keys.PAGE_DOWN).perform()

        col, data = get_salary()
        salary_data.append(data)

        time.sleep(2)

        year+=1
        cnt+=1

    
    return col, salary_data

        
        
```

```python
kbo_copy = merged_kbo_salary_h.copy()

def fill_mean():
    for index, row in kbo_copy.iterrows():
        zero_columns = row[row == 0].index.tolist()
        mean_val = kbo_copy.loc[(kbo_copy['연봉(만원)'] == row['연봉(만원)']) , zero_columns].mean().round(0)
        
        if len(mean_val)>0 and mean_val.values[0]==0:
            mean_val = kbo_copy[zero_columns].mean().round(0)

        kbo_copy.loc[index, zero_columns] = mean_val
   
```
